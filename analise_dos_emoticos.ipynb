```python
import os, emoji
from collections import Counter
```

```python
# Itera os arquivo de comentÃ¡rios e tranforma-os em uma lista de strings
# Pasta onde estÃ£o os arquivos

pasta_corpus = "corpus_comentarios_elis/"
lista_de_comentarios = []
# Para cada arquivo dentro de uma pasta
for arquivo in os.listdir(pasta_corpus):
    if arquivo.endswith(".txt"):
        caminho_arquivo = os.path.join(pasta_corpus, arquivo)

        # Abre e lÃª o arquivo
        with open(caminho_arquivo, "r", encoding="utf-8") as f:
            comentario = f.read().strip()  # Remove espaÃ§os extras
            lista_de_comentarios.append(comentario)
```

```python
# verifica as ocorrÃªncias de um emoticon especÃ­fico no corpos
placar = 0

comentarios_com_o_emoticon = []
for i, comentario in enumerate(lista_de_comentarios):
    if "â™‚" in comentario:
        placar += 1
        comentarios_com_o_emoticon.append(comentario)
        print(f"{i} - {comentario}")
print(placar)
```

```python
# Verifica quantos comentÃ¡rios possuem ao menos um emoji

comentarios_com_emoticon = 0

for comentario in lista_de_comentarios:
    # Extrair todos os emojis de cada comentÃ¡rio
    emojis_no_comentario = [char for char in comentario if char in emoji.EMOJI_DATA]
    if emojis_no_comentario:
        comentarios_com_emoticon += 1

print(f"\nNÃºmero de comentÃ¡rios com pelo menos um emoticon: {comentarios_com_emoticon}")
```

```python
# Lista dos emoticons mais usados


# Contadores
frequencias = Counter()

for comentario in lista_de_comentarios:
    # Extrair todos os emojis de cada comentÃ¡rio
    emojis_no_comentario = [char for char in comentario if char in emoji.EMOJI_DATA]

    # Contagem simples dos emojis
    frequencias.update(emojis_no_comentario)

    # Se o comentÃ¡rio tiver pelo menos um emoji, conta como 1
    if emojis_no_comentario:
        comentarios_com_emoticon += 1

# Ordenando a frequÃªncia dos emoticons do maior para o menor
print("FrequÃªncia simples dos emoticons (ordenada do maior para o menor):")
for emoticon, count in frequencias.most_common():
    print(f"{emoticon}: {count}")
```
